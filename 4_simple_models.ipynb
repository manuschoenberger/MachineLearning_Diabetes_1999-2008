```python
from sklearn.feature_selection import SelectKBest
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB, BernoulliNB
import pandas as pd
import numpy as np
```

# Load training and validation set

```python
train = pd.read_csv("train_set.csv")
train
```

```python
validation = pd.read_csv("validation_set.csv")
validation
```

```python
X_train = train.drop("readmitted", axis=1)
y_train = train.readmitted
X_val = validation.drop("readmitted", axis=1)
y_val = validation.readmitted
```

```python
validation.readmitted.value_counts()
```

```python
train.readmitted.value_counts()
```

# Simple model
Create a model and select the best feature to analyse the results -> recall of the minority class is 0.0

```python
selector = SelectKBest(k=1)

selector.fit(X_train, y_train)
best_feature = selector.get_feature_names_out()
print(best_feature)
```

```python
clf = OneVsRestClassifier(LogisticRegression())
clf.fit(X_train[best_feature], y_train)
```

```python
y_pred = clf.predict(X_val[best_feature])
```

```python
np.unique(y_val, return_counts=True)
```

```python
print(classification_report(y_val, y_pred))
```

# Function to find the best hyperparameters for k features

```python
def get_best_hyperparams(X_train, y_train, param_grid, clf):
    cv = 5
    # Select K Best Features
    selector = SelectKBest(k=25)
    selector.fit(X_train, y_train)

    # Get the selected feature names
    best_features = selector.get_feature_names_out()

    # Use the best features for training
    train_features = X_train[best_features]
    print(train_features.shape)

    # Perform GridSearchCV to find the best parameters
    gridsearch = GridSearchCV(clf, param_grid, cv=cv, scoring='f1_macro')
    gridsearch.fit(train_features, y_train)
    
    # Print the best parameters and the best score
    print("Best params: ", gridsearch.best_params_, "Best score:", gridsearch.best_score_)
```

# Apply different supervised learning methods

## Logistic Regression

### 25 best features (k = 25)

```python
param_grid={"C":np.logspace(-3,3,7), "penalty":["l1","l2"], "class_weight": [None, 'balanced']}
get_best_hyperparams(X_train, y_train, param_grid, LogisticRegression(multi_class='multinomial'))
```

```python
selector = SelectKBest(k=25)
selector.fit(X_train, train.readmitted)
best_features = selector.get_feature_names_out()
clf = LogisticRegression(C=1000.0, class_weight='balanced', penalty='l2')
clf.fit(X_train[best_features], train.readmitted)

y_pred = clf.predict(X_val[best_features])
print(classification_report(validation.readmitted, y_pred))
```

# Decision Tree Classifier

```python
param_grid={
    'max_depth': [*np.arange(1,50, 1), None], 
    "criterion": ['gini', 'entropy']
}

get_best_hyperparams(X_train, y_train, param_grid, DecisionTreeClassifier())
```

```python
selector = SelectKBest(k=25)
selector.fit(X_train, train.readmitted)
best_features = selector.get_feature_names_out()
clf = DecisionTreeClassifier(criterion='entropy', max_depth=17)
clf.fit(X_train[best_features], train.readmitted)

y_pred = clf.predict(X_val[best_features])
print(classification_report(validation.readmitted, y_pred))
```

# Naive Bayes

Using a Naive Bayes classifier with this dataset is possible, but it may not be the best choice due to a few key factors:

**Naive Bayes Assumptions:** Naive Bayes assumes that features are independent and normally distributed (for Gaussian Naive Bayes), which is not the case for this dataset. This can lead to suboptimal performance.

**Imbalanced Target Classes:** Since the target variable has a strong imbalance (with the majority class being 'NO' and the minority class '<30 days'), Naive Bayes might struggle to handle this imbalance effectively.

## Gaussian Naive Bayes

```python
param_grid = {'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]}

get_best_hyperparams(X_train, y_train, param_grid, GaussianNB())
```

```python
# Select the best features using SelectKBest
selector = SelectKBest(k=25)
selector.fit(X_train, train.readmitted)
best_features = selector.get_feature_names_out()

# Train a Gaussian Naive Bayes classifier with the selected features
clf = GaussianNB(var_smoothing=1e-09)
clf.fit(X_train[best_features], train.readmitted)

# Predict on the validation set using the selected features
y_pred = clf.predict(X_val[best_features])

# Print the classification report
print(classification_report(validation.readmitted, y_pred))
```

## Bernoulli Naive Bayes

```python
param_grid = {'alpha': [0.1, 0.5, 1.0], 'binarize': [0.0, 0.5, 1.0]}

get_best_hyperparams(X_train, y_train, param_grid, BernoulliNB())
```

```python
# Select the best features using SelectKBest
selector = SelectKBest(k=25)
selector.fit(X_train, train.readmitted)
best_features = selector.get_feature_names_out()

# Train a Gaussian Naive Bayes classifier with the selected features
clf = BernoulliNB(alpha=0.5, binarize=0.0)
clf.fit(X_train[best_features], train.readmitted)

# Predict on the validation set using the selected features
y_pred = clf.predict(X_val[best_features])

# Print the classification report
print(classification_report(validation.readmitted, y_pred))
```
